# Project Title

<font size = 5>Webscraping Real Estate Websites for Housing Market Analysis</font>

**<font color = 'green' size=4>Project Summary</font>**
    
In this project we will scrape the **REMAX** website (https://www.remax.ca/) to get information of the housing properties up for sale in different cities across different provinces in Canada. We will be using **Selenium** to assist in scraping the **REMAX** for both the basic and detailed information of each listing across different pages. We will use this scraped data do the following:
    
- Map these listings to quickly identify regions based on their price
    
- Plot distributions of housing prices across different regions for comparison
    
- Visualize housing price distribution based on housing type (condos, single family etc.)
    
- Find average price of listing across different cities and identify most and least expensive cities
    
- Identify top 10 most expensive and top 10 least expensive listings
    
- Analyze *Price* vs *Square Footage* trend for each region and identify any similarities or discrepancies
   
**<font color = 'green' size=4>List of steps followed:</font>**
- **Step 1**: Gather basic and details data from **REMAX** website for various cities and save them locally
- **Step 2**: Append all the saved scraped data as one pandas dataframe
- **Step 3**: Merge basic and details data on unique #MLS key, perform EDA and treat missing data
- **Step 4**: Correct data types , extract new features and remove redundant or misleading data
- **Step 5**: Use geopy library to extract latitude and longitude coordinates using the address of each listings and add it to dataframe
- **Step 6**: Create a dashboard using **Plotly** and **Dash** to answer all the questions
- **Step 7**: Deploy the dashboard on **heroku** (https://housingmarketanalysis.herokuapp.com/)

# Combine Allthe Saved  Webscraping Data From Different Runs Together as a Single DataFrame

```python
#----------------------------------------------------
# Read all the saved webscraped data
#----------------------------------------------------

# Toronto
tor_basic_df  = pd.read_csv('Toronto/toronto_on_basic_df', index_col=0)
tor_basic_df['Region'] = 'Toronto'
tor_detail_df = pd.read_csv('Toronto/toronto_on_detail_df', index_col=0)

# Vancouver
van_basic_df  = pd.read_csv('Vancouver/vancouver_bc_basic_df', index_col=0)
van_basic_df['Region'] = 'Vancouver'
van_detail_df = pd.read_csv('Vancouver/vancouver_bc_detail_df', index_col=0)

# Ottawa
ott_basic_df  = pd.read_csv('Ottawa/ottawa_on_basic_df', index_col=0)
ott_basic_df['Region'] = 'Ottawa'
ott_detail_df = pd.read_csv('Ottawa/ottawa_on_detail_df', index_col=0)

# Hamilton
ham_basic_df  = pd.read_csv('Hamilton/hamilton_on_basic_df', index_col=0)
ham_basic_df['Region'] = 'Hamilton'
ham_detail_df = pd.read_csv('Hamilton/hamilton_on_detail_df', index_col=0)

# Calgary
cal_basic_df  = pd.read_csv('Calgary/calgary_ab_basic_df', index_col=0)
cal_basic_df['Region'] = 'Calgary'
cal_detail_df = pd.read_csv('Calgary/calgary_ab_detail_df', index_col=0)

# Halifax
hal_basic_df  = pd.read_csv('Halifax/halifax_ns_basic_df', index_col=0)
hal_basic_df['Region'] = 'Halifax'
hal_detail_df = pd.read_csv('Halifax/halifax_ns_detail_df', index_col=0)

#----------------------------------------------------------------
# Merge basic and details information from all scraped sources
#----------------------------------------------------------------
basic_df_sources = [tor_basic_df, van_basic_df, ott_basic_df ,
                    cal_basic_df, ham_basic_df, hal_basic_df]

details_df_sources = [tor_detail_df, van_detail_df, ott_detail_df ,
                      cal_detail_df, ham_detail_df, hal_detail_df]

# Merge basic dataframes
basic_df = pd.DataFrame()
for df in basic_df_sources:
    basic_df = basic_df.append(df)

# Merge details dataframes
details_df = pd.DataFrame()
for df in details_df_sources:
    details_df = details_df.append(df)
```

## Data Preprocessing

### Remove columns which have more than threshold percentage of data missing

```python
# Set the percentage threshold
threshold = 90

# Identify list of columns with missing percentage values above the set threshold
cols_below_threshold = (missing_by_cols[missing_by_cols
                          ['Missing_Percentage_Rounded'] > threshold].index)

# Drop the columns from the dataframe
remax.drop(cols_below_threshold, axis=1, inplace =True)
```

### Correct Data Types , Extract Features and Clean Data

```python
#------------------------------------------------------------------------------
# Trim all columns and Replace NaNs for room dimension columns
#------------------------------------------------------------------------------
remax = remax.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Replac NaNs with empty strings
remax['Bedroom'].fillna('', inplace =True)
remax['Dining Room'].fillna('', inplace =True)
remax['Living Room'].fillna('', inplace =True)
remax['Kitchen'].fillna('', inplace =True)

#------------------------------------------------------------------------------
# Add Extra Feature Columns
#------------------------------------------------------------------------------
# Add Province Name Column
remax['Province'] = remax['Address'].apply(lambda x:x.split(',')[-1].strip())

# Add City Name Column
remax['City'] = remax['Address'].apply(lambda x:x.split(',')[-2].strip())

# Add Street Name Column From Address
remax['Street'] = remax['Address'].apply(lambda x:x.split(',')[-3])
remax['Street'] = (remax['Street'].apply(lambda x:
                        (re.findall('(-?\d*.*\w*)',x)[0]).split('-')[-1].strip()))

# Create a Street_City column to later use for lat, long coordinates
remax['Street_City'] = remax['Street'] + ' , ' +remax['City']

# Extract Property Tax Amount from Property Tax attribute
remax['Property Tax'] = remax['Property Tax'].replace('N/A',np.NaN)
remax['Property Tax Amount'] =  (remax['Property Tax'].apply(lambda x:str(x).split('(')[0].
                                                        replace('$','').replace(',','').
                                                            strip()))

remax['Property Tax Amount'] = remax['Property Tax Amount'].astype(float)

# Add area features using custom functions
remax['Total_Sqft'] = cf.sqft_area(remax, 'Square Footage')
remax['Bedroom_Sqm'] = cf.area_from_dims(remax,'Bedroom')
remax['Dining_Sqm'] = cf.area_from_dims(remax,'Dining Room')
remax['Living_Sqm'] = cf.area_from_dims(remax,'Living Room')
remax['Kitchen_Sqm'] = cf.area_from_dims(remax,'Kitchen')

#------------------------------------------------------------------------------
# Convert data types and Drop redundant columns and Rows
#------------------------------------------------------------------------------
# Convert Price to Integer
remax['Price'] = remax['Price'].apply(lambda x:int(x.replace('$','').replace(',','')))

# Convert date objects to datetime
remax['Date Listed'] = pd.to_datetime(remax['Date Listed'])
remax['Last Updated'] = pd.to_datetime(remax['Last Updated'])

#------------------------------------------------------------------------------
# Column Operations
#------------------------------------------------------------------------------
# Keep only these relevant columns
filter_list = (['MLS#','Province','Region','City', 'Subdivision','Street','Address',
                'Street_City', 'Property Type','Date Listed', 'Price',
                'Beds', 'Baths', 'Total_Sqft', 'Bedroom_Sqm', 'Dining_Sqm',
                'Living_Sqm', 'Kitchen_Sqm',
                
                'Cooling', 'Heating','Parking', 'Basement',
                'Amenities', 'Features', 'Exterior Features'])

remax = remax[filter_list]

#------------------------------------------------------------------------------
# Row Operations
#------------------------------------------------------------------------------
# Drop Rows where Property Type is unkown or is a Vacant Land
remax = remax[remax['Property Type'].notnull()]
remax = remax[remax['Property Type']!='Vacant Land']

# Drop Rows where Beds are unknown
remax = remax[remax['Beds']!='n/a bed']

# Drop Anomalous Total_Sqft and Misleading House Values
remax.drop(remax.loc[remax['Total_Sqft']>30000].index, inplace=True)
remax.drop(remax.loc[remax['Price']<100000].index, inplace=True)
```

### Run <font color = 'green'> *latlong* </font> Function and Fetch Longitude and Latitude Information using Street and City Address and Save Final Cleaned DataFrame

run **help(cf.latlong)** to get more information on this function

<font color ='red' size=3> **NOTE: The following block of code takes a while to execute since it runs geopy functions on each and every address one at a time** </font> 

```python
#------------------------------------------------------------------------------
# Extract Lat Long using latlong function
#------------------------------------------------------------------------------
# Create LatLong column
remax['LatLong'] = remax['Street_City'].apply(lambda x:cf.latlong(x))

# Extract Langitude and Longitude From LatLong column
remax['Latitude'] = remax['LatLong'].apply(lambda x:x[0])
remax['Longitude'] = remax['LatLong'].apply(lambda x:x[1])

# Save the final cleaned DataFrame
remax.to_csv('remax_df.csv', index=False)
```

# Create a Dash App Using Plotly

**<font size =4> Run <font color = 'green'> *pe_8_grid_3cols_layout* </font> Function to Initialize and create the Dash App </font>**

run **help(pe_8_grid_3cols_layout)** to get more information on this function

```python
#---------------------------------------------------------------------------------
# Specify the filter controls properties and run app initialization
#---------------------------------------------------------------------------------
module_title = 'RealEstate Price Analysis'
dropdown_filter_property = 'Province'
dropdown_filter_property2 = 'Beds'
range_filter_property = 'Price'
checkbox_filter_property = 'Property Type'

# Create a dataframe 
df = pd.read_csv('remax_df.csv')

# Filter to Longitudes and Latitudes within Canada, remove wrong coordinates
df = df[(df['Longitude']<-40) & (df['Latitude']>40)]

app = pe_8_grid_3cols_layout(module_title, df, 
                             dropdown_filter_property,dropdown_filter_property2,
                             range_filter_property, checkbox_filter_property)


#---------------------------------------------------------------------------------
# Run the app and add visualizations
#---------------------------------------------------------------------------------
# Specify the input for callbacks and outputs to be retured
####################################################################################
@app.callback(
    [Output('fig1','figure'),
     Output('fig2','figure'),
     Output('fig3','figure'),
     Output('fig4','figure'),
     Output('fig5','figure'),
     Output('fig6','figure'),
     Output('fig7','figure'),],
    [Input('dropdown_id','value'),
     Input('dropdown_id2','value'),
     Input('range_id','value'),
     Input('checkbox_id','value')])

# Set up plots
#---------------------------------------------------------------------------------
# Function which runs and updates plots based on change of filters
#---------------------------------------------------------------------------------
def update_graph(dropdown_id, dropdown_id2, range_id, checkbox_id):
    dff = df[df[dropdown_filter_property].isin(list(dropdown_id))]
    dff = dff[dff[dropdown_filter_property2].isin(list(dropdown_id2))]
    dff = dff[(dff[range_filter_property] >= min(range_id)) & (dff[range_filter_property] <= max(range_id))]
    dff = dff[dff[checkbox_filter_property].isin(list(checkbox_id))]
    
    #---------------------------------------------------------------------------------
    # Create all the required plots to populate on callback
    #---------------------------------------------------------------------------------
    px.set_mapbox_access_token('pk.eyJ1Ijoid2NkLXBhcnJ5IiwiYSI6ImNrZ2F3emExaTA1MTMyem1xYWlxaTJmOXQifQ.uNDZBfkfSYD1JxGsW94fzg')
    fig1 = px.scatter_mapbox(dff, 'Latitude', 'Longitude',color='Region',size='Price',
                             template='custom',hover_name = 'Address',zoom=3)
    
    fig2 = px.violin(dff,'Region','Price',color='Region',template='custom', points = False,
                   title='Price by Region')
    fig2.update_layout(showlegend=False)
    fig2.update_yaxes(range=[0, 3e6])
    
    fig3 = px.box(dff, 'Price', 'Property Type', color='Property Type',template='custom', 
                  title = 'Price Distribution by Unit Type')
    fig3.update_layout(showlegend=False)
    fig3.update_xaxes(range=[0, 7e6])

    
    mean_price_city = dff.groupby(dff['City'].str.capitalize()).mean()[['Price']].sort_values(by='Price',
                                                                                             ascending=False)
    fig4 = px.bar(mean_price_city, 'Price', mean_price_city.index ,
                  color=mean_price_city.index,template='custom', 
                  title = 'Average Price by City')
    fig4.update_layout(showlegend=False)
    
    expensive_vs_cheap = dff.groupby(dff['Address']).mean()[['Price']].sort_values(by='Price')
    expensive = expensive_vs_cheap.tail(10)
    fig5 = px.bar(expensive, 'Price', expensive.index, color = 'Price', template='custom',
                  color_continuous_scale=px.colors.sequential.Viridis_r,
                  title = 'Ten Most Expensive Listings')

    cheap = expensive_vs_cheap.head(10)
    fig6 = px.bar(cheap, 'Price', cheap.index, color='Price', template='custom', 
                  color_continuous_scale=px.colors.sequential.Viridis_r,
                  title = 'Ten Least Expensive Listings')
    fig6.update_yaxes(autorange='reversed')

    fig7 = px.scatter(dff, 'Total_Sqft' , 'Price',color='Region', trendline='ols',
                      template='custom', 
                      title = 'Price vs Area Correlation')
    fig7.update_yaxes(range=[0, 20e6])

    return fig1,fig2, fig3, fig4, fig5, fig6, fig7
    

if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
